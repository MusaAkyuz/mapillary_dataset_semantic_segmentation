```python
import os
import PIL.Image as Image
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
from utils.definitions import *
```

```python
# Check the image and their label image
random_image_name = random.choice(os.listdir(TRAIN_IMAGE))
random_image_path = os.path.join(TRAIN_IMAGE, random_image_name)

random_label_name = random_image_name.replace(".jpg", ".png")
random_label_path = os.path.join(TRAIN_LABEL, random_label_name)

random_image =cv2.imread(random_image_path)
random_label = cv2.imread(random_label_path)

# Mask the ROAD_COLOR (128, 64, 128) to 1 and others to 0
mask_of_label = cv2.inRange(random_label, (128, 64, 128), (128, 64, 128))

plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.imshow(random_image)
plt.subplot(1, 3, 2)
plt.imshow(random_label)
plt.subplot(1, 3, 3)
plt.imshow(mask_of_label)
plt.show()
```
